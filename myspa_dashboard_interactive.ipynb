```python
import pandas as pd
import numpy as np
import plotly.express as px
from faker import Faker
import random

print("Ready to build!")
```

```python
fake = Faker('de_DE')
random.seed(42)

cities = ['Berlin', 'Munich', 'Frankfurt', 'Hamburg', 'Darmstadt']
services = ['Private Spa', 'Birthday Party', 'Couples Session', 'Group Session']
fnb_items = ['Pizza', 'Coffee', 'Coca Cola', 'Champagne', 'None']
months = list(range(1, 13))

num_bookings = 50000

data = {
    'booking_id': range(1, num_bookings + 1),
    'client_name': [fake.name() for _ in range(num_bookings)],
    'city': [random.choice(cities) for _ in range(num_bookings)],
    'month': [random.choice(months) for _ in range(num_bookings)],
    'service_type': [random.choice(services) for _ in range(num_bookings)],
    'fnb_order': [random.choice(fnb_items) for _ in range(num_bookings)],
    'group_size': [random.randint(1, 10) for _ in range(num_bookings)],
    'duration_hours': [random.choice([1, 2, 3, 4]) for _ in range(num_bookings)],
    'revenue_eur': [random.randint(80, 800) for _ in range(num_bookings)],
    'supply_cost_eur': [random.randint(20, 150) for _ in range(num_bookings)],
    'satisfaction_score': [random.randint(1, 10) for _ in range(num_bookings)],
}

df = pd.DataFrame(data)
df['profit_eur'] = df['revenue_eur'] - df['supply_cost_eur']
print(df.head())
```

```python
city_profit = df.groupby('city')['profit_eur'].sum().reset_index()

fig = px.bar(city_profit, 
             x='city', 
             y='profit_eur',
             title='Total Profit by City',
             color='city')
fig.show()
```

```python
city_profit = df.groupby('city')['profit_eur'].sum().reset_index()

fig = px.bar(city_profit, 
             x='city', 
             y='profit_eur',
             title='Total Profit by City',
             color='city')
fig.show()
```

```python
city_profit = df.groupby('city')['profit_eur'].sum().reset_index()

fig = px.bar(city_profit, 
             x='city', 
             y='profit_eur',
             title='Total Profit by City',
             color='city')

fig.write_html('city_profit.html')
print("Chart saved! Open city_profit.html to see it")
```

```python
monthly_revenue = df.groupby('month')['revenue_eur'].sum().reset_index()

fig2 = px.line(monthly_revenue,
               x='month',
               y='revenue_eur',
               title='Revenue Trend by Month',
               markers=True)

fig2.write_html('monthly_revenue.html')
print("Chart saved!")
```

```python
service_revenue = df.groupby('service_type')['revenue_eur'].sum().reset_index()

fig3 = px.pie(service_revenue,
              values='revenue_eur',
              names='service_type',
              title='Revenue Distribution by Service Type')

fig3.write_html('service_revenue.html')
print("Chart saved!")
```

```python
satisfaction = df.groupby('city')['satisfaction_score'].mean().reset_index()

fig4 = px.bar(satisfaction,
              x='city',
              y='satisfaction_score',
              title='Average Customer Satisfaction by City',
              color='city')

fig4.write_html('satisfaction.html')
print("Chart saved!")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=['Total Profit by City', 
                    'Revenue Trend by Month',
                    'Revenue by Service Type',
                    'Customer Satisfaction by City'])

fig.add_trace(go.Bar(x=city_profit['city'], y=city_profit['profit_eur'], name='Profit'), row=1, col=1)
fig.add_trace(go.Scatter(x=monthly_revenue['month'], y=monthly_revenue['revenue_eur'], name='Revenue', mode='lines+markers'), row=1, col=2)
fig.add_trace(go.Bar(x=service_revenue['service_type'], y=service_revenue['revenue_eur'], name='Service'), row=2, col=1)
fig.add_trace(go.Bar(x=satisfaction['city'], y=satisfaction['satisfaction_score'], name='Satisfaction'), row=2, col=2)

fig.update_layout(title='MySpa Germany - Complete Business Dashboard', height=700)
fig.write_html('complete_dashboard.html')
print("Dashboard saved!")
```
